##Guardando un Notebook en GitHub desde Colaboratory

El siguiente paso es conocer como guardar Notebooks elaborados desde Colaboratory en un repositorio de GitHub.


Accese a su cuenta de google y abra su primer Notebook desde Colaboratory, donde vera una ventana como se muestra en la siguiente imagen:

2020-08-25 (5).png

Acccione el menu **FILE** y seleccione la opcion **Save a copy in GitHub**, se abrira una ventana emergente para validar su usuario y clave de acceso a GitHub y luego se desplegara otra como podra observar en la siguiente imagen:

2020-08-25 (6).png

En primer lugar se debe indicar en cual de sus repositorios de GitHub desea guardar el Notebook, para este momento seleccionaremos el repositorio creado especialente para este diplomado.

Seguidamente se debe indicar cual es la rama (Branch) donde va a pertenecer el Notebook.

El siguiente campo es para indicar la ruta de acceso dentro del repositorio, para este caso bastara con el nombre del Notebook (MiPrimerNotebook) con su extencion (.ipynb).

Luego encontramos un campo donde debemos colocar un mensaje. Este mensaje es muy importante cada vez que se guarda un Notebook, pues permite indicar a primera vista una idea clara acerca de este envio de informacion a su repositorio. Esto es de ayuda cuando se trabaja en equipo con colaboradores y cuando deben hacerse revisiones del trabajo en macha.

Por ultimo active la casilla de verificacion, esto agregara un enlace directo desde esta copia del archivo hacia la version original en su Colaboratory.

Una vez hecho accione donde se indica **ok** para guardarlo, eso desplegara una nueva ventana como la siguiente imagen, donde podra observar el Notebook almacenado en el repositorio desde la plataforma de GitHub. Note el enlace azul que indica **Open in Colab** en color blanco, accionando ese boton podra dirigirse directamente a su Notebook en Colaboratory.

> Pero el acceso no sera publico porque google tendra que validar si usted configuro el permiso de acceso a el usuario en particular que acciono el boton.

Ahora bien, si desea ofrecer acceso publico es preferible copiar desde el navegador la direccion completa a este Notebook. Note que esta resaltada en azul con texto de color blanco en esta imagen.

##Revisando un Documento en un Repositorio

Un ejemplo de esto podemos observarlo en la siguiente imagen, donde se presentan las ultimas dos versiones de un documento. Note que son resaltados los cambios hechos, donde en rojo se presenta como estava en la version anterior y en verde como esta en la ultima version.
